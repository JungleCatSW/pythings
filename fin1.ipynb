```python
import pandas as pd

# Select a date range
start_date = '2017-02-01'
end_date = '2018-01-01'
dates = pd.date_range(start_date,end_date)
# create a dataframe with exisitng dates
df1 = pd.DataFrame(index=dates)


# INDEX IS AN INTEGER CONVERT TO DATE
dfSPY = pd.read_csv("SPY.csv",
                    index_col="Date",
                    parse_dates=True,
                    usecols=['Date','Adj Close'],
                    na_values=['nan'])
# Use join to combine stocks
dfSPY = dfSPY.rename(columns={'Adj Close': 'SPY'})
df1 = df1.join(dfSPY, how='inner') 
# join only rows common to both data frames
# this is the same as dropping nans  ... df1.dropna()
#print("data frame")
#print(df1)

# rad in many stocks
symbols = ['NVDA','AAPL','GOOG']
for symbol in symbols:
    df_temp = pd.read_csv(
        "{}.csv".format(symbol),
        index_col='Date',
        parse_dates=True,
        usecols=['Date','Adj Close'],
        na_values=['nan'])
    df_temp = df_temp.rename(columns={"Adj Close": symbol})
    #df_temp = df_temp.rename({'Adj Close', symbol}, axis='columns')
    #print(df_temp)
    
    df1 = df1.join(df_temp)
print(df1)

```

## Slicing a Data Frame

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
# Select a date range
start_date = '2017-02-01'
end_date = '2018-01-01'
dates = pd.date_range(start_date,end_date)
# create a dataframe with exisitng dates
df1 = pd.DataFrame(index=dates)


# INDEX IS AN INTEGER CONVERT TO DATE
dfSPY = pd.read_csv("SPY.csv",
                    index_col="Date",
                    parse_dates=True,
                    usecols=['Date','Adj Close'],
                    na_values=['nan'])
# Use join to combine stocks
dfSPY = dfSPY.rename(columns={'Adj Close': 'SPY'})
df1 = df1.join(dfSPY, how='inner') 

symbols = ['NVDA','AAPL','GOOG']
for symbol in symbols:
    df_temp = pd.read_csv(
        "{}.csv".format(symbol),
        index_col='Date',
        parse_dates=True,
        usecols=['Date','Adj Close'],
        na_values=['nan'])
    
    df_temp = df_temp.rename(columns={"Adj Close": symbol})
    df1 = df1.join(df_temp)
    #ix deprecated

print(df1.loc['2017-02-01':'2017-02-08'])
print(df1.loc['2017-02-01':'2017-02-08',['SPY','NVDA']])
#plot stocks
ax = df1.plot(title='Stock Prices')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
plt.show()

```
